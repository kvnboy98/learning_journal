# Data Preprocessing

```python
data_inf = data.sample(10, random_state=17)
data_inf
```

```python
data_train_test =  data.drop(data_inf.index)
data_train_test 
```

```python
data_train_test.reset_index(drop=True, inplace=True)
data_inf.reset_index(drop=True, inplace=True)
```

```python
data_inf
```

## Splitting Dataset

```python
X = data_train_test.drop(columns='Rating')
y = data_train_test['Rating']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
#test_size adalah 80 persen untuk training dan 20 persen untuk tes
print(X_train.shape)
print(X_test.shape)
```

## Handling Missing Value

## Feature Selection

```python
X_train
```

karena `nama` tidak ada hubunganya dengan rating maka bisa diasumsikan bahwa kolom ini dihilangkan.

```python
X_train.drop(columns='Name', inplace=True)
X_test.drop(columns='Name', inplace=True)
```

```python
X_test.head()
```

## split kolom numerik dan kategorikal

```python
num_columns = X_train.select_dtypes(include=np.number).columns.to_list()
cat_columns = X_train.select_dtypes(include='object').columns.to_list()
print(num_columns)
print(cat_columns)
```

```python
X_train_num = X_train[num_columns]
X_train_cat = X_train[cat_columns]

X_test_num = X_test[num_columns]
X_test_cat = X_test[cat_columns]
```

```python
X_train_num.head()
```

```python
X_train_cat.head()
```

## Feature Scalling

```python
X_train_num
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train_num)
#difit train dulu biar tahu rentang setiap kolomnya

X_train_num_scaled = scaler.transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
X_train_num_scaled
```

----
## Feature Encoding

```python
X_test_cat
```

```python
from sklearn.preprocessing import OrdinalEncoder
# kalau punya tingkatan (hi n low) pake ordinalencoder, kalau gk punya tingkatan ONEHOT ENCODER
encoder = OrdinalEncoder()
encoder.fit(X_train_cat)

X_train_cat_enc = encoder.transform(X_train_cat)
X_test_cat_enc = encoder.transform(X_test_cat)
```

```python
X_test_cat_enc
```

## Concat numerik kolom dan kategorikal

```python
X_train_final = np.concatenate([X_train_num_scaled, X_train_cat_enc], axis=1)
X_test_final = np.concatenate([X_test_num_scaled, X_test_cat_enc], axis=1)
```

```python
X_train_final
```

```python
x_train_final_df = pd.DataFrame(X_train_final, columns=[num_columns + cat_columns])
x_train_final_df
```

# Model Definition

```python
from sklearn.linear_model import LinearRegression
model_lin_reg = LinearRegression()
```

metrics yang dipakai MAE (MEAN ABSOLUTE ERROR)

# Model Training

Cell pada bagian ini hanya berisi code untuk melatih model dan output yang dihasilkan. Lakukan beberapa kali proses training dengan hyperparameter yang berbeda untuk melihat hasil yang didapatkan. Analisis dan narasikan hasil ini pada bagian Model Evaluation.

```python
model_lin_reg.fit(X_train_final, y_train)
```

# Model Evaluation

```python
y_pred_train = model_lin_reg.predict(X_train_final)
y_pred_test = model_lin_reg.predict(X_test_final)
y_pred_test
```

```python
from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)

print('error_train = ', mae_train)
print('error_test = ', mae_test)
```
