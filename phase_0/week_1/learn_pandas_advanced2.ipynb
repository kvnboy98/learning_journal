# MENGGANTI MISSING VALUE YG BUKAN NaN

```python
df
```

```python
cnt=0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        print('ini angka')
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        print('ini bukan angka')
        #pass
    cnt+=1
```

```python
df
```

```python
cnt=0
for row in df['NUM_BATH']:
    try:
        float(row)
        print('ini angka')
    except ValueError:
        df.loc[cnt, 'NUM_BATH']=np.nan
        print('ini bukan angka')
        #pass
    cnt+=1
```
