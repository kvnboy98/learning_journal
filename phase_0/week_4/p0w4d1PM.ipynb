```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.plot()
```

```python
x=np.arange(0,150,1)
y=x^2-x
```

```python
fig, axes = plt.subplots(ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes = axes.flatten()
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,y)
axes[3].plot(y,x)
```
